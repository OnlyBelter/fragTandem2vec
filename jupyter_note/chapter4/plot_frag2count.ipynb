```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
from tqdm import tqdm
import matplotlib.pyplot as plt
```

### model mol2vec

```python
frag2num_mol2vec = pd.read_csv('./frag2num_all_moses_model_mol2vec_no_remove.csv')
frag2num_mol2vec.head(2)
```

```python
frag2num_mol2vec.shape
```

```python
total_frag_mol2vec = frag2num_mol2vec['count'].sum()
total_frag_mol2vec
```

```python
plt.figure(figsize=(8, 6))
plt.hist(np.log2(frag2num_mol2vec['count']), bins=25, label='fragment by fingerprints')
plt.xlabel('$log_2$(Fragment count in MOSES dataset)')
plt.ylabel('Number of fragment')
plt.legend()
plt.tight_layout()
plt.savefig('model_mol2vec_plot_frag2count.png', dpi=200)
```

### model parallelFragment

```python
frag2num_parallelFrag = pd.read_csv('./frag2num_all_moses_model_parallelfragment.csv')
frag2num_parallelFrag.head(2)
```

```python
frag2num_parallelFrag.shape
```

```python
total_frag_parallelFrag = frag2num_parallelFrag['count'].sum()
total_frag_parallelFrag
```

```python
plt.figure(figsize=(8, 6))
plt.hist(np.log2(frag2num_parallelFrag['count']), bins=25, label='fragment by tree decomposition')
plt.xlabel('$log_2$(Fragment count in MOSES dataset)')
plt.ylabel('Number of fragment')
plt.legend()
plt.tight_layout()
# plt.savefig('model_parallelFrag_plot_frag2count.png', dpi=200)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(np.log2(frag2num_parallelFrag['count']), bins=25, density=True, label='fragment by tree decomposition', histtype='step')
plt.hist(np.log2(frag2num_mol2vec['count']), bins=25, density=True, label='fragment by fingerprints', histtype='step')

plt.xlabel('$log_2$(Fragment count in MOSES dataset)')
plt.ylabel('PMF')
plt.legend()
plt.tight_layout()
plt.savefig('compare_two_models_plot_frag2count.png', dpi=200)
```
