```python
import pandas as pd
import numpy as np
```

```python
cid2ps = pd.read_csv('../big-data/purity_score/mol2vec/cid2class_with_purity_score_0.csv')
```

```python
cid2ps['batch'] = 0
```

```python
cid2ps.head()
```

```python
cid2ps_list = []
for i in range(10):
    cid2ps_fp = '../big-data/purity_score/mol2vec/cid2class_with_purity_score_{}.csv'.format(i)
    cid2ps = pd.read_csv(cid2ps_fp)
    cid2ps['batch'] = i
    cid2ps_list.append(cid2ps)
```

```python
cid2ps_all = pd.concat(cid2ps_list)
print(cid2ps_all.shape)
cid2ps_all.head()
```

```python
pure_cluster = cid2ps_all[cid2ps_all['purity_score']>=0.7].copy()
pure_cluster.head()
```

```python
total_pure_cluster_count = pure_cluster.groupby(['batch'])['total'].sum()
total_pure_cluster_count
```

```python
np.mean(total_pure_cluster_count), np.std(total_pure_cluster_count)
```

```python
pure_cluster.groupby(['batch', 'superclass_name']).count()
```

```python
pure_score = pure_cluster.groupby(['batch'])['purity_score'].mean()
pure_score
```

```python
np.mean(pure_score), np.std(pure_score)
```
